# Recommendation system 

```python
# Importing necessary modules
from colorama import Fore, Style  # Importing colorama for colored text in terminal 
from surprise import Dataset, Reader, SVD  # Importing necessary modules from surprise library
from surprise.model_selection import train_test_split  # Importing train_test_split function

# Initializing a Reader object with the rating scale
reader = Reader(rating_scale=(1, 5))

# Retrieving unique values of the 'Review Rating' column from the DataFrame df and sorting them
unique_rating = sorted(df['Review Rating'].unique())


# Loading data from DataFrame into a Surprise Dataset object
surprise_data = Dataset.load_from_df(df[['Customer ID', 'Item Purchased', 'Review Rating']], reader)

# Splitting the dataset into training and testing sets
trainset, testset = train_test_split(surprise_data, test_size=0.2, random_state=1)

# Creating an instance of the SVD (Singular Value Decomposition) model
model = SVD()

# Training the model on the training set
model.fit(trainset)
# Infinite loop to repeatedly ask for customer identification code
while True:
    try:
        # Prompting the user to enter their identification code
        customer_identification_code = int(input('Enter your ID please: '))

        # Checking if the entered identification code is in the DataFrame's 'Customer ID' column
        if customer_identification_code in df['Customer ID'].unique():
            # Exiting the loop if the identification code is valid
            break
        else:
            # Printing a message indicating invalid ID if the identification code is not found in the DataFrame
            print(f"{Fore.YELLOW}INVALID ID{Style.RESET_ALL}")
    except ValueError:
        # Handling the ValueError exception if the input cannot be converted to an integer
        print(f"{Fore.RED}Invalid input. Please enter a valid integer.{Fore.RED}")
        
```

```python
# Extracting the list of products purchased by a specific customer using their identification code
products_user_rated = df[df['Customer ID'] == customer_identification_code]['Item Purchased'].tolist()

# Obtaining the list of all unique products available
all_products = df['Item Purchased'].unique()

# Finding the products that the user has not rated yet
unrated_products = set(all_products) - set(products_user_rated)

# Generating predictions for the user's ratings for the unrated products using a model
user_predictions = [model.predict(customer_identification_code, product) for product in unrated_products]

# Prompting the user to input the number of product recommendations they desire
top_n = int(input('Enter how many product recommendations you want: '))

# Selecting the top n recommended products based on estimated ratings
recommended_products = sorted(user_predictions, key=lambda x: x.est, reverse=True)[:top_n]

# Extracting the IDs of the recommended products
recommended_product_ids = [product.iid for product in recommended_products]
# Printing the top recommended products for the customer with their identification code
print(f"\nTop{Fore.GREEN} {top_n} recommended products for Customer {customer_identification_code}: {recommended_product_ids}")
```
